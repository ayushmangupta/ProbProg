#### Normal Distribution

```python
mean = [0,0,0,0,0,0]
std = [.05,.1,.5,1,2,5]
for k, N in enumerate(std):
    sx = plt.subplot(len(std)/2, 2, k+1)

    plt.style.use('fivethirtyeight')
    plt.title("Normal Distribution")
    
    y = stats.norm.pdf(x,loc= mean[k],scale=std[k])

    plt.plot(x, y, label="Mean %d ,\n Std %d " % (mean[k], std[k]))
    #plt.fill_between(x,.24,y, color="#348ABD", alpha=0.4)
    leg = plt.legend()
    leg.get_frame().set_alpha(0.4)
    plt.autoscale(tight=True)
    plt.grid(True)

    
plt.tight_layout()
```

## Beta Distribution

```python
dist = stats.beta
n_trials = [0, 1, 2, 3, 4, 5, 8, 15, 50, 500]
data = stats.bernoulli.rvs(0.5, size=n_trials[-1])
x = np.linspace(0, 1, 100)


for k, N in enumerate(n_trials):
    plt.style.use('fivethirtyeight')
    sx = plt.subplot(len(n_trials)/2, 2, k+1)
    plt.xlabel("$p$, probability of heads") \
        if k in [0, len(n_trials)-1] else None
    plt.setp(sx.get_yticklabels(), visible=False)
    heads = data[:N].sum()
    y = dist.pdf(x, 1 + heads, 1 + N - heads)
    plt.plot(x, y, label="observe %d tosses,\n %d heads" % (N, heads))
    plt.fill_between(x, 0, y, color="#348ABD", alpha=0.4)
    plt.vlines(0.5, 0, 4, color="k", linestyles="--", lw=1)

    leg = plt.legend()
    leg.get_frame().set_alpha(0.4)
    plt.autoscale(tight=True)
    plt.grid(True)


plt.suptitle("Bayesian updating of posterior probabilities",
             y=1.02,
             fontsize=14)

plt.tight_layout()
```

## Exponential Distribution

$$Formula:    \lambda e^{-\lambda x} $$
$$ Mean:  1/\lambda$$ $$  STD: 1/{\lambda^2}$$  
$$ \lambda >0,  x = [0, \infty] $$


```python
x = np.linspace(0,15)
x
```

```python
def expo(l,x):
    return l*np.exp(-l*x)
```

```python
fig, ax = plt.subplots(figsize=(5,7))
plt.plot(x,expo(.1,x),label = '0.1',lw=2)
plt.plot(x,expo(.5,x),label = '0.5',lw=2)
plt.plot(x,expo(.7,x),label = '0.8',lw=2)
plt.plot(x,expo(.9,x),label = '0.9',lw=2)
plt.legend()
```

```python
init_data = 1
def compound(init_data,years,rate):
    var = [init_data]
    for i in range(1,years+1):
        var.append(np.round(var[-1]*(1+rate),decimals=2))
    return var

def simple(init_data,years,rate):
    var = [init_data]
    for i in range(1,years+1):
        var.append(np.round(init_data*(1+i*rate),decimals=2))
    return var
```

```python
years=20
interest_rate = .15
initial_investment = 100
x = list(range(0,years+1))
y1 = compound(initial_investment,years,interest_rate)
y2 = simple(initial_investment,years,interest_rate)
texts_comp = [str(x) for x in y1]
texts_simp = [str(x) for x in y2]
fig, ax = plt.subplots(figsize=(12,9))
plt.plot(range(years+1),y1,lw='1',label="Compound Interest")
plt.plot(range(years+1),y2,lw='1',label="Simple Interest");
ax.xaxis.set_ticks(np.arange(0, 20+1,1));
ax.yaxis.set_ticks(np.arange(0,max(compound(initial_investment,years,interest_rate)) ,100));
plt.legend()
for i,tx in enumerate(texts_comp):
    ts = texts_simp[i]
    xx = x[i]
    yy = y1[i]
    yy2 = y2[i]
    plt.scatter(xx, yy, marker='o', color='red')
    plt.text(xx+.3, yy+20, tx, fontsize=9)
    plt.scatter(xx, yy, marker='o', color='red')
    plt.text(xx+.3, yy2-20, ts, fontsize=9)
    #plt.scatter(x, y, marker='x', color='red')
    #plt.text(x+0.3, y+0.3, type, fontsize=9)


plt.legend();
```
